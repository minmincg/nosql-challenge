```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo=MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments=db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
pprint(db.establishments.find_one())
```

```python
# Find the establishments with a hygiene score of 20
hygiene = {'scores.Hygiene':{'$eq':20}}
# Use count_documents to display the number of documents in the result
print("Number of documents with hygiene score equal to 20:",establishments.count_documents(hygiene))
# Display the first document in the results using pprint
pprint(db.establishments.find_one(hygiene))
```

```python
hygiene_data = establishments.find(hygiene)
hygiene_data
```

```python
# Convert the result to a Pandas DataFrame
hygiene20_df=pd.DataFrame(hygiene_data)
```

```python
# Display the number of rows in the DataFrame
print(len(hygiene20_df))
# Display the first 10 rows of the DataFrame
hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london = {'LocalAuthorityName':{'$regex':"London"},'RatingValue':{'$gte':'4'}}

# Use count_documents to display the number of documents in the result
print("Number of documents in London:",establishments.count_documents(london))

# Display the first document in the results using pprint
pprint(db.establishments.find_one(london))
london_data=db.establishments.find(london)
```

```python
# Convert the result to a Pandas DataFrame
london_df=pd.DataFrame(london_data)
# Display the number of rows in the DataFrame
print(len(london_df))
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

nearest = {'geocode.latitude':{'$lte':(latitude+degree_search),'$gte':(latitude-degree_search)},
           'geocode.longitude':{'$lte':(longitude+degree_search),'$gte':(longitude-degree_search)},
           'RatingValue':{'$gte':'5'}
          }
sort =  [('scores.Hygiene',-1)]
# Print the results
pprint(list(establishments.find(nearest).sort(sort)))
nearest_data=db.establishments.find(nearest)
```

```python
# Convert result to Pandas DataFrame
nearest_df=pd.DataFrame(nearest_data)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query={'$match':{'scores.Hygiene':{'$eq':0}}}
# 2. Groups the matches by Local Authority
group_query={'$group':{'_id':"$LocalAuthorityName",'count':{'$sum':1}}}
# 3. Sorts the matches from highest to lowest
sort_values={'$sort':{'count':1}}

pipeline=[match_query,group_query,sort_values]
local_authority_data=list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print("Number of documents in the result: ",len(local_authority_data))
# Print the first 10 results
pprint(local_authority_data[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_df=pd.DataFrame(local_authority_data)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(local_authority_data))
# Display the first 10 rows of the DataFrame
local_authority_df.head(10)
```
